メイン処理：ファイルの読み込み単語の長さを求め保存

```python
word_lengths = []
with open('/content/drive/MyDrive/nlpseminar2024/simple_corpus.txt') as f:
  for data_line in f:
    data_line = data_line.rstrip()
    words = data_line.split(' ')# 入力文を単語に分割
    for word in words:
      l = len(word)             # len()で文字列の長さ（文字数）
      word_lengths.append(l)    # word_lengthsに長さlを格納

print(word_lengths)
```

ヒストグラムを描画

```python
import matplotlib.pyplot as plt # 描画用モジュールの読み込み

# ヒストグラムの描画
plt.hist(word_lengths, bins=20) # 区間は20個
plt.show()  # 画面表示
```
