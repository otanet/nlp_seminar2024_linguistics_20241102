メイン処理：ファイルの読み込み単語の頻度をカウント

```python
word2frequency = {} #辞書型変数の宣言

with open('/content/drive/MyDrive/dl/data/simple_corpus.txt' ) as file:
  for data_line in file:
    data_line = data_line.rstrip()
    tokens = analyzer(data_line) # 入力文を解析（含むトークナイズ）
    for token in tokens:
      print(word2frequency)
      word2frequency[token.text] = word2frequency.get(token.text, 0) + 1 # 単語を一つづつカウント
```

カウント結果を出力

```python
for word, frequency in word2frequency.items():
  print(word, frequency)
```
